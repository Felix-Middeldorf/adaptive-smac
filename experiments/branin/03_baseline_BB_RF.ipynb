1. Create Bramin SMAC Blackboxfacade run with Random Forest Surrogate

```python
import numpy as np

def branin(x1, x2):
    a = 1.0
    b = 5.1 / (4 * np.pi**2)
    c = 5 / np.pi
    r = 6.0
    s = 10.0
    t = 1 / (8 * np.pi)

    return a * (x2 - b * x1**2 + c * x1 - r)**2 + s * (1 - t) * np.cos(x1) + s
```

```python
from ConfigSpace import ConfigurationSpace, Configuration, Float
from smac import Scenario, BlackBoxFacade, AlgorithmConfigurationFacade

SEED = 67
N_TRIALS = 100

def branin_eval(config: Configuration, seed: int = 0):
    x1 = config["x1"]
    x2 = config["x2"]
    return branin(x1,x2)

cs = ConfigurationSpace(seed=SEED)
cs.add(
    Float( "x1", (-5,10)),
    Float( "x2", (0,15))
)

sc = Scenario(
    configspace=cs,
    deterministic=True,
    n_trials=N_TRIALS,
    seed=SEED
)

# pull the Random forest from AC
model = AlgorithmConfigurationFacade.get_model(scenario = sc)

# create Blackboxfacade with ACFacade-RF surrogate
smac = BlackBoxFacade(
    scenario=sc,
    target_function=branin_eval,
    model = model
    #overwrite=True
)

incumbent = smac.optimize()
print(incumbent)
print(smac.runhistory.get_cost(incumbent))
```
